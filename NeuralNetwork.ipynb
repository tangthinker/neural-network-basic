```python
import scipy.special
```

```python
class NeuralNetwork:
    # 设定输入、隐藏、输出节点数量
    def __init__(self,input_size, hidden_size, output_size, learning_rate):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.learning_rate = learning_rate

        # self.W_input_hidden = (numpy.random.rand(self.hidden_size, self.input_size) - 0.5)
        # self.W_hidden_output = (numpy.random.rand(self.output_size, self.hidden_size) - 0.5)
        # 使用正太分布创建初始权重矩阵
        self.W_input_hidden = numpy.random.normal(0.0, pow(self.hidden_size, -0.5), (self.hidden_size, self.input_size))
        self.W_hidden_output = numpy.random.normal(0.0, pow(self.hidden_size, -0.5), (self.output_size, self.hidden_size))
        self.activation_func = lambda x : scipy.special.expit(x)
        self.inverse_activation_func = lambda x : scipy.special.logit(x)
        pass
    # 学习给定训练集样本，优化权重
    def train(self, inputs, targets):
        inputs = numpy.array(inputs, ndmin=2).T
        targets = numpy.array(targets, ndmin=2).T

        X_hidden = numpy.dot(self.W_input_hidden, inputs)
        O_hidden = self.activation_func(X_hidden)

        X_output = numpy.dot(self.W_hidden_output, O_hidden)
        O = self.activation_func(X_output)

        E_output = targets - O
        E_hidden = numpy.dot(self.W_hidden_output.T, E_output)
        self.W_hidden_output += self.learning_rate * numpy.dot(E_output * O * (1.0 - O), O_hidden.T)
        self.W_input_hidden += self.learning_rate * numpy.dot(E_hidden * O_hidden * (1.0 - O_hidden), inputs.T)
        pass
    # 给定输入，预测输出
    def query(self, inputs):
        inputs = numpy.array(inputs, ndmin=2).T
        X_hidden = numpy.dot(self.W_input_hidden, inputs)
        O_hidden = self.activation_func(X_hidden)

        X_output = numpy.dot(self.W_hidden_output, O_hidden)
        O = self.activation_func(X_output)
        return O
        pass
    def back_query(self, targets):
        O = numpy.array(targets, ndmin=2).T
        
        X_output = self.inverse_activation_func(O)
        
        O_hidden = numpy.dot(self.W_hidden_output.T, X_output)
        
        O_hidden -= numpy.min(O_hidden)
        O_hidden /= numpy.max(O_hidden)
        O_hidden *= 0.98
        O_hidden += 0.01
        
        X_hidden = self.inverse_activation_func(O_hidden)
        
        inputs = numpy.dot(self.W_input_hidden.T, X_hidden)
        inputs -= numpy.min(inputs)
        inputs /= numpy.max(inputs)
        inputs *= 0.98
        inputs += 0.01
        
        return inputs
    
```

```python
input_size = 3
hidden_size = 3
output_size = 3

learning_rate = 0.3

nn = NeuralNetwork(input_size, hidden_size, output_size, learning_rate)
```

```python
training_date_file = open("minist_data/mnist_train.csv", 'r')
training_date_list = training_date_file.readlines()
training_date_file.close()
```

```python
for record in training_date_list:
    all_values = record.split(',')
    inputs = (numpy.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01
    targets = numpy.zeros(output_size) + 0.01
    targets[int(all_values[0])] = 0.99
    nn.train(inputs, targets)
```

```python
score_card=[]

for record in test_date_list:
    all_values = record.split(',')
    label = int(all_values[0])
    print("correct label:", label)
    test_inputs = (numpy.asfarray(all_values[1:]) / 255.0 * 0.99 ) + 0.01
    outputs = nn.query(test_inputs)
    O_label = numpy.argmax(outputs)
    print("network's answer:", O_label)
    if (label == O_label):
        score_card.append(1)
    else:
        score_card.append(0)    
```

```python
score_arr = numpy.asarray(score_card)
score = score_arr.sum() / score_arr.size
score
```
